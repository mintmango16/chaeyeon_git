## 당뇨병 데이터_의사결정 트리 분석 

```python
import pandas as pd
```

```python
 # 데이터 로드
data = pd.read_csv("./Pima Indians Diabetes Database/diabetes.csv") # 상대경로 지정 
data
```

**데이터 feature 정보**
- Pregnancies : 임신 횟수 
- Glucose : 경구 포도당 내성 검사에서 2시간 후의 혈장 포도당 농도
- BloodPressure : 이완기 혈압(mmHg)
- SkinThickness : 삼두근 피부 주름 두께(mm)
- Insulin : 2시간 혈청 인슐린(mu U/ml)
- BMI : 체질량지수(체중(kg)/키(m)^2)
- DiabetesPedigreeFunction : 당뇨병 혈통 기능 
- Age : 나이
- Outcome : 당뇨병 진단 유무,  768개 중 268개가 1(True)이고 나머지는 0(False)

### **데이터 분리**

```python
X = data.drop(["Outcome"], axis = 1) # 종속변수 제거
y = data["Outcome"] # 종속변수만 저장 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### **의사결정 트리 모델 생성 및 학습**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree1 = DecisionTreeClassifier()
#tree = DecisionTreeClassifier(max_depth=2) # 트리의 최대 깊이 
tree1.fit(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(tree1, out_file = "diabetes1.dot", feature_names = X.columns, 
               class_names = ['0','1'], rounded=True, filled=True, impurity=True)
```

```python
import pydot

graph1 = pydot.graph_from_dot_file("diabetes1.dot")[0]

iris_png1 = graph1.create_png()
```

```python
from IPython.core.display import Image 

Image(iris_png1)
```

```python
y_pred = tree1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred) # 정확도 확인 
```

---

```python
tree2 = DecisionTreeClassifier(max_depth=2) # 트리의 최대 깊이 
tree2.fit(X_train, y_train)
```

```python
export_graphviz(tree2, out_file = "diabetes2.dot", feature_names = X.columns, 
               class_names = ['0','1'], rounded=True, filled=True, impurity=True)
```

```python
graph2 = pydot.graph_from_dot_file("diabetes2.dot")[0]

iris_png2 = graph2.create_png()
```

```python
Image(iris_png2)
```

```python
y_pred = tree2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred) # 정확도 확인 
```
