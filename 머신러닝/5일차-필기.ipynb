```python
### 지니계수 계산 방식
- 경제적 불평등의 정도를 계수화한 것
- 불순도 확인 ( 불순한 정도 )
- 특정 사건이 일어날 확률 -> yes_prob
- 특정 사건이 일어나지 않을 확률 -> no_prob
- A = yes_prob * ( 1 - yes_prob )
- B = no_prob * ( 1 - no_prob )
- A + B = gini
```

```python
size = 19798

yes_prob = 5509 / size # 일어날 확률
no_prob = 14289 / size # 일어나지 않을 확률

A = yes_prob * ( 1 - yes_prob )
B = no_prob * ( 1 - no_prob )

print(A + B)   # 0.402 = 지니계수 
```

```python
yes_prob = 0
no_prob = 1

A = yes_prob * ( 1 - yes_prob )
B = no_prob * ( 1 - no_prob )

print(A + B) 
```

### 알고리즘 구분 용어 ( 대부분이 ensemble, bagging, boosting ) 임
- ensemble : 음악에서 연주할 때 합창, 협주 -> 1인 X, 같이 결과를 도출하는 것 (RandomForest - 여러개의 tree를 도출 )
- 회귀문제 : n개의 알고리즘 결과값(숫자)들의 평균
- 분류분제 : n개의 알고리즘 결과값(경우의 수)들이 투표
- 대표적인 알고리즘 : Random Forest(부스팅 계열), Extra Trees, Gradient Boostong)
- bagging  
- boosting

### bagging
- bootstrap + aggregation + ing : b + agg + ing
- bootstrap: 데이터 셋의 row를 무작위로 복원 추출 -> 새로운 데이터셋 생성 ( 중복된 row가 있을 수 있음 ) - 여러개 생성
- train  : test > **train 데이터 중에서 복원 추출** > 분류 : 투표, 회귀: 평균

- 동일한 데이터와 알고리즘으로 학습하면 항상 같은 결과가 나옴 -> bootstrap을 이용하면 새로운 데이터셋을 만들 때마다 row의 구성이 조금씩 다르게 됨(무작위로 복원 추출하기 때문에)
- bootstrap으로 서로 조금씩 다른 데이터셋을 n개 만들고, n개의 데이터셋에 대해 학습은 진행(조금 다르게 학습됨)  -> n개의 학습된 예측 결과를 합산, 병렬 연산 가능

### boosting : Tunning을 병행해야 함. 
#### - 병렬 연산이 불가능하기 때문에 충분한 시간을 가지고 해야함. 그렇지 않으면 RandomForest 추진

- 오답을 이용해 저수 상승을 도모, 오답을 새로운 정보로 이용한 접근법 통칭
- 첫 번째 알고리즘으로 학습하고 예측하여 오답 정보를 만듦 > 오답 정보의 문제를 학습대상 데이터셋에 추가해 새로운 데이터셋 생성

## 혼동 행렬 (Confusion Matrix)

|  | Predicted class | P  | N  |
|-------------|----------------|----|----|
| Actual class | |  |  |
|-------------|----------------|----|----|
| **P**       |       | TP | FN |
| **N**       |       | FP | TN |


### 기호 설명
- **P** : 예측 결과가 예약 취소인 경우
- **N** : 예측 결과가 예약 취소가 아닌 경우
- **T** : 실제 결과와 비교 시 정답인 경우
- **F** : 실제 결과와 비교 시 오답인 경우

### 평가 지표
- **TP (True Positive)** : 예약 취소됨을 맞춘 경우
- **TN (True Negative)** : 예약 취소 아님을 맞춘 경우
- **FP (False Positive)** : 예약 취소 아닌데 예약 취소로 분류한 경우 (잘못된 경고)
- **FN (False Negative)** : 예약 취소인데 예약 취소 아님으로 분류한 경우 (놓친 예약 취소)



### 분류 문제의 평가 지표

#### **accuracy(정확도)**
- 예측결과와 실제값이 맞은 건수 / 전체 데이터 수
- acc = TP + TN / TP + TN + FN + FP

#### **precision(정밀도)**
- 모델이 **True**라고 분류한 것 중에서 실제 **True**인 것의 비율
- prec = TP / (TP + FP)
- 날씨 예측 모델이 맑다고 예측했는데, 실제 날씨가 맑았는지 살펴보는 지표

#### **recall(재현율)**

- 실제 True인 것 중에서 모델이 True라고 예측한 것의 비율
- recall = TP / (TP + FN)
- 실제 날씨가 맑은 날 중에서 모델이 맑다고 예측한 비율을 나타낸 지표
- sensitivity, hit rate

#### 언제 accuracy 대신에 precision과 recall 등이 고려 될까 ?
- 클래스 불균형 문제가 발생할 때 ACCURACY의 의미가 떨어짐

- True(90%) 인 경우와 False(10%)인 경우 분류   - 잘 못 맟춤 ( class imbalance issue )

#### fall-out
- 실제 False인 데이터 중에서 모델이 True라고 예측한(잘 못 예측한) 비율
- fall-out = FP / ( FP + TN )

#### Specificity ( 특이도 )
- 실제 False인 데이터 중에서 모델이 False라고 예측한 비율
- spec = TN / ( FP + TN )

### F1-Score
- 정밀도와 재현율을 결함 지표
- 정밀도와 재현율의 차이가 없을 때 높은 값을 가짐

- F1 = (2 recall * spec ) / (recal + precision)

### Precision, Recall의 차이점

#### 화재 경보 ( 불이 난 경우 : 1, 불이 안난 경우 : 0 )

- FP 는 불이 났다고 판단했으나 불이 안 남 -> False Alram
- FN 은 불이 안 났다고 판단했으나 불이 남 -> Miss   -> FN에 비중(중점)을 주어야 함.(발생하면 안되는 현상)

#### 스팸 메일 분류 ( 스팸인 경우 : 1, 아닌 경우 :  0)
- FP는 스팸이라고 판단했으나 스팸이 아님
- FN은 스팸이라고 판단했으나 스팸임
- 중요한 메일이 오기로 했는데 스팸으로 분류되면 확인 못함 -> FP에 중점

#### 환자 검진 (병에 걸린 경우 : 1, 아닌 경우 : 0 )
- FP는 병에 걸렸다고 판단했는데 이상이 없음
- FN는 병에 걸리지 않았다고 판단했으나 병이 있음
- FP에 중점을 두면 다행이지만 환자의 불란감이 증가, FN에 중점을 두면 큰일

### 게임 어뷰저 분류(어뷰저인 경우 : 1, 아닌 경우: 0 ) : 불법적으로 게임을 하는 사용자 경우

```python
from sklearn.metrics import confusion_matrix
```

```python
TN, FP, FN, TP = confusion_matrix(for_plot['actual'], for_plot['predict']).ravel()
```

```python
print(TN, FP, FN, TP)
```

```python
# Accuracy (정확도)
acc = (TP + TN ) / (TN + FP +  FN + TP)
print( "Accuracy : ", acc)
```

```python
# precision (정밀도)
prec = TP / (TP + FP)
print( "Precision : ", prec)
```

```python
# Recall(재현율)

recall = TP / (TP + FN)
print(" Recall : ", recall)
```

```python
# fall-out
fall_out = FP / ( FP + TN )
print("Fall-out : ", fall_out)
```

```python
# Specificity ( 특이도 )

spec = TN / ( FP + TN )
print("Specificity : ", spec)
```

```python
# F1-Score

F1 = (2 * recall * spec ) / (recall + prec)
print("F1-score : ", F1)
```

### Hyperparameter Tuning
- Hyperparameter : 사용자가 지정해야만 하는 parameter
- 최적의 옵션값을 찾아 알고리즘 성능을 극대화 시키는 방법

```python
model = rf()
model.fit(X_train, y_train)

for_plot = pd.DataFrame()
for_plot['predict'] = model.predict(X_test)
for_plot['actual'] = y_test
```

```python
model = rf(max_depth=10, max_features=0.9)
model.fit(X_train, y_train)

for_plot = pd.DataFrame()
for_plot['predict'] = model.predict(X_test)
for_plot['actual'] = y_test

TN, FP, FN, TP = confusion_matrix(for_plot['actual'], for_plot['predict']).ravel()

recall = TP / ( TP + FN )
print(recall)
```

```python
def scorer(model):
    model = rf(max_depth=10, max_features=0.9)
    model.fit(X_train, y_train)
    
    for_plot = pd.DataFrame()
    for_plot['predict'] = model.predict(X_test)
    for_plot['actual'] = y_test
    
    TN, FP, FN, TP = confusion_matrix(for_plot['actual'], for_plot['predict']).ravel()
    
    recall = TP / ( TP + FN )

    return recall
```

```python
model = rf(max_depth=12, max_features=0.7)
scorer(model)
```

### 접근벙 3 가지
- Grid-Search : 격자(방충망)
- random-Search : 무작위 조합
- Optimizer

```python
md_ls = [3, 5, 7, 9, 11, 13, 15, 17]
mf_ls = [0.1, 0.3, 0.5, 0.7, 0.9]
all_scores = []

for md in md_ls:
    for mf in mf_ls:
        model = rf(max_depth=md, max_features=mf, n_jobs=-1)
        all_scores.append({'recall' : scorer(model), 'md': md, 'mf': mf})

hyper_df = pd.DataFrame(all_scores)
```

```python
hyper_df.sort_values(by='recall', ascending=False)
```
