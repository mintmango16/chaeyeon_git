# feature selection
- 과적합을 방지하고, 일반화를 달성하기 위해 사용하는 방법
- filter, wrapper, embed 

Feature Selection은 주어진 데이터에서 중요한 변수를 선택하고, 덜 중요한 변수를 제거하여 모델의 성능과 해석 가능성을 향상시키는 과정입니다. 
과적합을 방지하고, 일반화를 달성(90~95% 달성)하기 위해 사용하는 방법
이 과정은 데이터 전처리 단계에서 수행되며, 선택된 변수는 다음 단계의 모델 학습에 사용됩니다

## Feature Selection의 필요성

- 과적합 방지: 불필요한 변수(관련 없는 변수, 다중 공선성 변수)를 제거함으로써 모델이 불필요 패턴을 학습하는 것을 방지
- 모델 성능 향상: 노이즈를 줄이고 중요한 정보만을 사용하여 모델의 정확도를 높입니다.
 - 독립 변수 10~15개만 되어도 충분
- 계산 비용 감소: 변수가 적으면 학습 및 예측 시간이 단축됩니다.
- 모델 해석 용이성: 중요한 변수만으로 구성된 모델은 해석하기가 더 쉽습니다.

## Feature Selection의 유형

### 필터 방법(Filter Methods):

- 통계적 지표(예: 분산, 상관계수)를 기반으로 중요한 변수를 선택합니다.

- 예 : Chi-square Test, ANOVA, Mutual Information.

### 랩퍼 방법(Wrapper Methods):

- 특정 머신러닝 모델을 활용하여 변수의 중요성을 평가합니다.

- 예: Forward Selection, Backward Elimination, Recursive Feature Elimination (RFE).

### 임베디드 방법(Embedded Methods):

- 모델 학습 과정에서 변수의 중요도를 계산하여 선택합니다.

- 예: Lasso Regression, Random Forest Feature Importance

## Feature Selection 방법론

### 필터 방법 예제: Variance Threshold

- 변수 간 분산이 낮은 경우 제거합니다.

### 랩퍼 방법 예제: Recursive Feature Elimination (RFE)

- 모델을 반복적으로 학습하면서 중요도가 낮은 변수를 제거합니다.

### 임베디드 방법 예제: Lasso Regression

- L1 정규화(규제)를 통해 가중치가 0이 되는 변수를 제거합니다.

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# 데이터 로드
housing = fetch_california_housing()
data = pd.DataFrame(housing.data, columns=housing.feature_names)
data['TARGET'] = housing.target

# 데이터 확인
print(data.head())

#열 9개 : 빈곤율/집 내구력/ 방 개수 평균/ 침실 개수 평균/인구수...
```

```python
from sklearn.feature_selection import VarianceThreshold  # 필터 방법

# 분산 기준 설정
threshold = 0.1 # 분산 기준
selector = VarianceThreshold(threshold=threshold)  #필터링의 기준 설정 : 분산 기준 설정
data_selected = selector.fit_transform(data) # 필터링 진행하여 다른변수에 저장

print(f"선택된 변수 수: {data_selected.shape[1]}")  # 필터링된 데이터 변수 수 확인
```

```python
# 랩퍼 방식 : Recurcive Deature Elimination(RFE)
from sklearn.feature_selection import RFE  #필터링 방법
from sklearn.linear_model import LinearRegression #로지스틱 회귀 모델

# 모델 설정 및 RFE 적용
model = LinearRegression()
rfe = RFE(estimator=model, n_features_to_select=5) # 필터링할 변수를 5개로 입력, estimator로 로지스틱 회귀 모델 사용
rfe.fit(data.iloc[:,:-1], data['TARGET']) # x 데이터 : data에서 슬라이싱으로 target 제외

# 션택된 변수 출력
selected_features = data.columns[:-1][rfe.support_]
print(f"선택된 변수: {selected_features.tolist()}") # 필터링된 데이터 변수 종류 확인
```

```python
#임베디드 방식 feature 분류 : Lasso(L1) 방식 : L1 정규화를 통해 가중치가 0이되는 변수 제거

from sklearn.linear_model import Lasso

# 모델 학습

lasso = Lasso(alpha=0.1)
lasso.fit(data.iloc[:,:-1], data['TARGET'])

# 중요도 기반 변수 선택 : 임베디드 변수 선택
lasso_selected = data.columns[:-1][lasso.coef_ != 0]  # 기울기가 0이 아닌 변수 필터
print(f"선택된 변수 : {lasso_selected.tolist()}")
```
